# Model #1 - First Order Regression Model with Quantitative and Qualitative Variables

You have been asked to create a first order regression model for *price* as the response variable, and *sqft_living*, *sqft_above*, *age*, *bathrooms*, and *view* as predictor variables. Before writing any code, review Section 3 of the Summary Report template to see the questions you will be answering about your first order multiple regression model.

Run your scripts to get the outputs of your regression analysis. Then use the outputs to answer the questions in your summary report.

**Note: Use the + (plus) button to add new code blocks, if needed.**

```r
plot(housing$price, housing$sqft_living,
    main="Price vs sqft_living",
     xlab="Price", ylab="Living Square Ft",
     xlim=c(0,1500000),
     ylim=c(0,6000),
     col="red",
     pch=19, frame=FALSE)
```

```r
plot(housing$price, housing$age,
    main="Price vs Age",
     xlab="Price", ylab="Age",
     xlim=c(0,1500000),
     ylim=c(0,150),
     col="red",
     pch=19, frame=FALSE)
```

```r
housing_subset_vars <- c("price", "age", "sqft_living")
housing_subset <- housing[housing_subset_vars]
head(housing_subset)
corr_matrix1 <- cor(housing_subset, method="pearson")
round(corr_matrix1,4)
```

```r
model1 <- lm(price ~ sqft_living + sqft_above + age + bathrooms + view, data=housing)
summary(model1)
```

```r
fitted_values1 <- fitted.values(model1)
residuals1 <- residuals(model1)

plot(fitted_values1, residuals1, 
    main="Residuals vs Fitted Values",
    xlab="Fitted Values", ylab="Residuals",
    col="red", pch=19)
```

```r
qqnorm(residuals1, pch=19, col="blue", frame= FALSE)
qqline(residuals1, col="red", lwd = 2)
```

```r
#prediction and conf intervals 1
newDF1 <- data.frame(sqft_living=2150, sqft_above=1050, age=15, bathrooms=3,view='0')

print("Prediction Interval")
predict_int1 <- predict(model1, newDF1, interval="predict", level=0.90)
round(predict_int1,4)

print("Confidence Interval")
conf_int1 <- predict(model1, newDF1, interval="confidence", level=0.90)
round(conf_int1,4)
```

```r
#prediction and conf intervals 2
newDF1 <- data.frame(sqft_living=4250, sqft_above=2100, age=5, bathrooms=5,view='2')

print("Prediction Interval 2")
predict_int1 <- predict(model1, newDF1, interval="predict", level=0.90)
round(predict_int1,4)

print("Confidence Interval 2")
conf_int1 <- predict(model1, newDF1, interval="confidence", level=0.90)
round(conf_int1,4)
```

# Model #2 - Complete Second Order Regression Model with Quantitative Variables

You have been asked to create a complete second order regression model for *price* as the response variable, and *school_rating* and *crime* as predictor variables. Before writing any code, review Section 4 of the Summary Report template to see the questions you will be answering about your complete second order multiple regression model.

Run your scripts to get the outputs of your regression analysis. Then use the outputs to answer the questions in your summary report.

**Note: Use the + (plus) button to add new code blocks, if needed.**

```r
print(range(housing$school_rating))
print(range(housing$crime))
```

```r
plot(housing$price, housing$school_rating,
    main="Price vs School_rating",
     xlab="Price", ylab="School_rating",
     xlim=c(0,1500000),
     ylim=c(0,15),
     col="red",
     pch=19, frame=FALSE)
```

```r
plot(housing$price, housing$crime,
    main="Price vs Crime",
     xlab="Price", ylab="Crime",
     xlim=c(0,1500000),
     ylim=c(0,500),
     col="red",
     pch=19, frame=FALSE)
```

```r
model2 <- lm(price ~ school_rating + crime + school_rating:crime + I(school_rating^2) + I(crime^2), data=housing)
summary(model2)
```

```r
fitted_values2<- fitted.values(model2)
residuals2<- residuals(model2)

plot(fitted_values2, residuals2,
    main="Residuals vs Fitted Values(Model2)",
    xlab="Fitted Values", ylab="Residuals",
    col="red", pch=19)
```

```r
qqnorm(residuals2, pch=19, col="blue", frame= FALSE)
qqline(residuals2, col="red", lwd = 2)
```

```r
#prediction and conf intervals 3 for model 2
newDF3 <- data.frame(school_rating=9.80, crime=81.02)

print("Prediction Interval")
predict_int3 <- predict(model2, newDF3, interval="predict", level=0.90)
round(predict_int3,4)

print("Confidence Interval")
conf_int3 <- predict(model2, newDF3, interval="confidence", level=0.90)
round(conf_int3,4)
```

```r
#prediction and conf intervals 4 for model 2
newDF4 <- data.frame(school_rating=4.28, crime=215.50)

print("Prediction Interval")
predict_int4 <- predict(model2, newDF4, interval="predict", level=0.90)
round(predict_int4,4)

print("Confidence Interval")
conf_int4 <- predict(model2, newDF4, interval="confidence", level=0.90)
round(conf_int4,4)
```
